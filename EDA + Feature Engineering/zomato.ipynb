```python
final_df = pd.merge(df,df_country,on='Country Code',how='right')
final_df.head()
```

```python
final_df.info()
```

```python
final_df.describe()
```

```python
final_df.shape
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0: 'Rating'})
ratings
```

```python
mpl.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating",data=ratings)
plt.legend()
plt.show()
```

```python
sns.barplot(x="Aggregate rating",y="Rating",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
