```python
q2 = np.median(data['loan_amnt'])
q1 = np.percentile(data['loan_amnt'], 25, method='midpoint')
q3 = np.percentile(data['loan_amnt'], 75, method='midpoint')


iqr = q3 - q1

lowerthresold = q1 - 1.5 * iqr
upperthresold = q3 + 1.5 * iqr

print('Q1',q1)
print('Q2',q2)
print('Q3',q3)
print('IQR',iqr)
print('lower thresold',lowerthresold)
print('upper thresold',upperthresold)
```

```python
outlier = []
for x in data['loan_amnt']:
    if x < lowerthresold or x > upperthresold:
        outlier.append(x)
outlier
```

```python
out = pd.DataFrame(outlier)

out
```

```python
final_data =  pd.merge(data, out, left_index=True, right_index=True)
final_data
```

```python
final_data =  pd.merge(data, out, on = 'None', how='outer')
final_data
```
