```python
missing_df = pd.DataFrame({
    "missing_count": df.isnull().sum(),
    "missing_percentage": (df.isnull().sum() / len(df)) * 100
})
missing_df
```

```python
missing_df = missing_df.sort_values(by="missing_percentage", ascending=False)

print(missing_df)


```

```python
plt.figure(figsize=(10, 6))
sns.barplot(
    x=missing_df["missing_percentage"],
    y=missing_df.index,
    palette=["red" if x > 30 else "blue" for x in missing_df["missing_percentage"]]
)
plt.xlabel("Missing Percentage")
plt.ylabel("Features")
plt.title("Missing Value Percentage by Feature")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(df["Age"].dropna(), kde=True, ax=axes[0])
axes[0].set_title("Age Distribution")

sns.histplot(df["Fare"], kde=True, ax=axes[1])
axes[1].set_title("Fare Distribution")

plt.show()


```

```python
def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = series[(series < lower) | (series > upper)]
    return outliers

age_outliers = detect_outliers(df["Age"].dropna())
fare_outliers = detect_outliers(df["Fare"])

print(f"Age Outliers: {len(age_outliers)} ({len(age_outliers)/len(df)*100:.2f}%)")
print(f"Fare Outliers: {len(fare_outliers)} ({len(fare_outliers)/len(df)*100:.2f}%)")

```

```python

plt.figure(figsize=(10, 4))
sns.boxplot(data=df[["Age", "Fare"]])
plt.title("Boxplot for Age & Fare")
plt.show()
```

```python
pivot_table = df.pivot_table(
    values="Survived",
    index="Sex",
    columns="Pclass",
    aggfunc="mean"
)

print(pivot_table)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.barplot(x="Sex", y="Survived", data=df, ax=axes[0])
axes[0].set_title("Survival Rate by Gender")

sns.barplot(x="Pclass", y="Survived", data=df, ax=axes[1])
axes[1].set_title("Survival Rate by Passenger Class")

plt.show()
```

```python
num_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10, 6))
sns.heatmap(num_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
df["IsAlone"] = np.where(df["FamilySize"] == 1, 1, 0)

family_survival = df.groupby("FamilySize")["Survived"].mean()
alone_survival = df.groupby("IsAlone")["Survived"].mean()

print(family_survival)
print(alone_survival)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.barplot(x=family_survival.index, y=family_survival.values, ax=axes[0])
axes[0].set_title("Survival Rate by Family Size")

sns.barplot(x=alone_survival.index, y=alone_survival.values, ax=axes[1])
axes[1].set_title("Survival Rate: Alone vs Not Alone")

plt.show()
```

```python

corr_target = num_df.corr()["Survived"].drop("Survived").sort_values(ascending=False)

print("Top 3 Correlated Features with Survival:")
print(corr_target.head(3))
```

```python

stronger_feature = "Fare" if abs(corr_target["Fare"]) > abs(corr_target["Age"]) else "Age"
print(f"Stronger predictor of Survival: {stronger_feature}")
```
