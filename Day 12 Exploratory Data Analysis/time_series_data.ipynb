```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller
```

```python
data = pd.read_csv("NIFTY 50.csv")
print(data.head())
```

```python
data.head()
```

RESAMPLING THE DATA

```python
High = pd.to_datetime(data['High'])
data = data.set_index('High')
data_resampled = data.resample("ME",on="High").mean(numeric_only=True)

sns.set_style='White'
plt.figure(figsize=(12,6))
sns.lineplot(
    data = data_resampled,
    x = data_resampled.index,
    y = 'High',
)
plt.show()
```
