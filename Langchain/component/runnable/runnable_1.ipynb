```python
import random
```

```python
# dummy component
class Dummy:
    
    def __init__(self):
        print("Dummy component initialized")
    
    def predict(self, prompt):
        response_list = [
            "This is a dummy response.",
            "The dummy component is working.",
            "Hello from the dummy component!"
        ]
        return {'response': random.choice(response_list)}
```

```python
llm = Dummy()
```

```python
llm.predict("What is the dummy component?")
```

```python
class DummyPromptTemplate:
    
    def __init__(self, template, input_variables):
        self.template = template
        self.input_variables = input_variables
        
    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
template = DummyPromptTemplate(
    template = "Write  poem on {topic}",
    input_variables=['topic'])
```

```python
prompt = template.format({'topic': 'nature'})
```

```python
prompt
```

```python
llm.predict(prompt)
```

chaining

```python
class DummyLLMChain:
    
    def __init__(self, llm, prompt):
        self.llm = llm
        self.prompt = prompt
        
    def run(self, input_dict):
        final_prompt = self.prompt.format(input_dict)
        result = self.llm.predict(final_prompt)
        return result['response']
```

```python
template = DummyPromptTemplate(
    template = "Write  poem on {topic}",
    input_variables=['topic'])
```

```python
chian = DummyLLMChain(llm=llm, prompt=template)
```

```python
chian.run({'topic': 'nature'})
```
