```python
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import (train_test_split,
                                      StratifiedKFold,
                                        LeaveOneOut,
                                        KFold,
                                        cross_val_score,
                                         cross_validate)
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier 
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR
from sklearn.datasets import load_iris
```

```python
km = KFold(n_splits=5, shuffle=True, random_state=42)
stf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
loo = LeaveOneOut()
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model_km = cross_val_score(RandomForestClassifier(), X, y, cv=km)
rf_model_sft = cross_val_score(RandomForestClassifier(), X, y, cv=stf)
rf_model_loo = cross_val_score(RandomForestClassifier(), X, y, cv=loo)
```

```python
print(rf_model_km.mean())
print(rf_model_sft.mean())
print(rf_model_loo.mean())
```

```python
dt_model_km = cross_val_score(DecisionTreeClassifier(), X, y, cv=km)
dt_model_sft = cross_val_score(DecisionTreeClassifier(), X, y, cv=stf)
dt_model_loo = cross_val_score(DecisionTreeClassifier(), X, y, cv=loo)
```

```python
print(dt_model_km.mean())
print(dt_model_sft.mean())
print(dt_model_loo.mean())
```

```python
from xgboost import XGBClassifier
xg_model_km = cross_val_score(XGBClassifier(), X, y, cv=km)
xg_model_sft = cross_val_score(XGBClassifier(), X, y, cv=stf)
xg_model_loo = cross_val_score(XGBClassifier(), X, y, cv=loo)
```

```python
print(xg_model_km.mean())
print(xg_model_sft.mean())
print(xg_model_loo.mean())
```
