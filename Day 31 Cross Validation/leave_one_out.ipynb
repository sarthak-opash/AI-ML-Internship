```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import LeaveOneOut, cross_val_score
from sklearn.datasets import load_iris
import numpy as np
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
loo = LeaveOneOut()
```

# XGBoost

```python
from xgboost import XGBClassifier
result = cross_val_score(XGBClassifier(), X, y, cv=loo)
print(f"Mean Accuracy Score: {result.mean()*100}%")
```

# Random Forest

```python
result = cross_val_score(RandomForestClassifier(n_estimators=10, max_depth=5, n_jobs=-1, random_state=42), X, y, cv=loo)
print(f"Mean Accuracy Score: {result.mean()*100}%")
```

```python
result = cross_val_score(RandomForestClassifier(n_estimators=5, max_depth=5, n_jobs=-1, random_state=42), X, y, cv=loo)
print(f"Mean Accuracy Score: {result.mean()*100 :.2f}%")
```

```python
result = cross_val_score(RandomForestClassifier(), X, y, cv=loo)
print(f"Mean Accuracy Score: {result.mean()*100 :.2f}%")
```

```python
from sklearn.model_selection import cross_validate
cv = cross_validate(RandomForestClassifier(), X, y, cv=5,
    scoring={
        "accuracy": "accuracy",
    })
cv
```
