```python

import re
from datetime import datetime

def convert_time(match):
    time_str = match.group().upper()
    try:
        dt = datetime.strptime(time_str, "%d/%m/%y, %I:%M %p")
        return dt.strftime("%d/%m/%y, %H:%M")
    except ValueError:
        return match.group()

with open("WhatsApp Chat with IT Squad.txt", "r", encoding="utf-8") as f:
    text = f.read()

pattern = r"\d{2}/\d{2}/\d{2}, \d{1,2}:\d{2}\s*[aApP][mM]"

new_text = re.sub(pattern, convert_time, text)

with open("output_.txt", "w", encoding="utf-8") as f:
    f.write(new_text)

print("Time format converted to 24-hour successfully!")

```

```python
f = open("Output_.txt", 'r', encoding='utf-8')
data = f.read()
print(data)
```

```python
patterns = r"\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s"
```

```python
messages = re.split(patterns, data)
print(messages)
```

```python
dates = re.findall(patterns, data)
dates
```

```python
df = pd.DataFrame({'user_messages': messages[1:], 'message_date': dates})

df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %H:%M - ')

df.rename(columns={'message_date': 'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python

users = []
messages = []

for message in df['user_messages']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_messages'], inplace=True)

df.head()
```

```python
import datetime as dt

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['hours'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
df.head()
```
