```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
import torch
```

```python
model_name = 't5-large'
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)
```

```python
input_text = "summarize: The old grandfather clock in the hallway had not ticked in twenty years, yet Arthur heard it chime thirteen times. He rushed downstairs, heart pounding, to find his cat, Jasper, sitting perfectly still on the grandfather clock's casing. Jasper looked at him, winked, and pointed a paw toward the dusty floorboards. There, tucked under a loose plank, was a leather journal he hadn’t seen since the incident. He opened it to the last entry: The secret is safe.Confused, Arthur looked back, but Jasper was gone, and the clock was silent again. The air felt colder, and the silence seemed to whisper, Almost."
inputs = tokenizer(input_text, return_tensors="pt")
```

```python
summary_ids = model.generate(inputs.input_ids, max_length=100, num_beams=1, early_stopping=False)

output_text = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print("Summary:", output_text)
```
