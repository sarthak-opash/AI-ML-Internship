```python
quality_counts = data['quality'].value_counts()
plt.figure(figsize=(8,5))
plt.bar(quality_counts.index, quality_counts, color='deeppink')
plt.title('Count Plot of Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
sns.set_style("darkgrid")
numerical_column = data.select_dtypes(include=["int64", "float64"]).columns

plt.figure(
    figsize=(14, len(numerical_column)*3)
)

for idx,feature in enumerate(numerical_column, 1):
    plt.subplot(len(numerical_column), 1, idx)
    sns.histplot(data[feature], kde=True)
    plt.title(f"{feature} | Skewness: {round(data[feature].skew(), 2)}")

plt.tight_layout()
plt.show()
```

```python
plt.figure(
    figsize= (
        10,6
    )
)

sns.swarmplot(x='quality',y='alcohol',data=data,palette='hsvm')
plt.title('Swarm Plot for Quality and Alcohol')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()
```

```python
sns.set_palette("Pastel1")

plt.figure(figsize=(10, 6))

sns.pairplot(data)

plt.suptitle('Pair Plot for DataFrame')
plt.show()
```

```python
data['quality'] = data['quality'].astype(str)
plt.figure(figsize=(10,8))
sns.violinplot(
    x="quality",
    y="alcohol",
    data=data,
    palette={
        '3': 'lightcoral', 
        '4': 'lightblue', 
        '5': 'lightgreen', 
        '6': 'gold',
        '7': 'lightskyblue',
        '8': 'lightpink'
    },
    alpha=0.7
)
plt.show()
```

```python
sns.boxplot(x='quality', y='alcohol', data=data)
```

```python
plt.figure(figsize=(15, 10))

sns.heatmap(data.corr(), annot=True, fmt='.2f', cmap='Pastel2', linewidths=2)

plt.title('Correlation Heatmap')
plt.show()
```
