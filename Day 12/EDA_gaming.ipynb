```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('online_gaming_behavior_insights.csv')
df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df[['Age']].count()
```

```python
le = LabelEncoder()
df['encoded'] = le.fit_transform(df[['Age']])
df

```

```python
pp_cols = ['Age', 'PlayTimeHours', 'SessionsPerWeek', 'AvgSessionDurationMinutes', 'PlayerLevel', 'EngagementLevel']
sns.pairplot(df[pp_cols], hue='EngagementLevel', corner=True, plot_kws={'alpha':0.6, 's':20})
plt.suptitle('Pairplot of selected numeric features (by EngagementLevel)', y=1.02)
plt.show()
```
