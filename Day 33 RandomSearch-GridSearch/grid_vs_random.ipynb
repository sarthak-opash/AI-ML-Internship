```python
from numpy.random import uniform
from numpy.random import randint
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
import pandas as pd 
import numpy as np 
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python

data1 = pd.read_csv('indian_engineering_student_placement.csv')
data2 = pd.read_csv('placement_targets.csv')
```

```python
data = pd.merge(data1, data2, on='Student_ID')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

cols = ['gender', 'branch', 'part_time_job', 'family_income_level', 'city_tier', 'internet_access', 'extracurricular_involvement', 'placement_status']

for col in cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
```

```python
data
```

```python
data['extracurricular_involvement'] = data['extracurricular_involvement'].fillna(data['extracurricular_involvement'].mean())
data
```

```python
data.isnull().sum()
```

```python
X = data.drop(columns=['placement_status', 'salary_lpa'])
y = data['placement_status']
```

```python
data.corr()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

# Parameter Selection For these data

```python
parameters = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7, 9],
    'learning_rate': [0.01, 0.1, 0.2, 0.3],
    'subsample': [0.7, 0.12]
}
```

```python

from scipy.stats import randint, uniform

param_random = {
    'n_estimators': randint(50, 100),  
    'max_depth': randint(3, 11),            
    'learning_rate': uniform(0.01, 0.3),  
    'subsample': uniform(0.6, 0.4)  
}
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

treemodel_grid = XGBClassifier()
cv=GridSearchCV(treemodel_grid,param_grid=parameters,cv=5,scoring='accuracy')
cv.fit(X_train,y_train)
```

```python
cv.best_params_
```

```python
cv.best_score_
```

```python
cv.best_index_
```

```python
treemodel_rand = XGBClassifier()
cv_rand=RandomizedSearchCV(treemodel_rand, param_distributions=param_random, cv=5, scoring='accuracy')
cv_rand.fit(X_train,y_train)
```

```python
cv_rand.best_params_
```

```python
cv_rand.best_score_
```

```python
cv_rand.best_index_
```

* For this specific data we can obserbe best score of both Search method is almost same so we will check efficient method by Accuracy Score by given relevant Parameter value

GridSearchCV Param

```python
treemodel = XGBClassifier(
    learning_rate=0.1,
    max_depth=9,
    n_estimators=50,
    subsample=0.7,
)
```

```python
treemodel.fit(X_train,y_train)
```

```python
y_pred = treemodel.predict(X_test)
accuracy_score(y_test, y_pred)*100
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

For RandomizedSearch

```python
treemodel_rand = XGBClassifier(
    learning_rate=0.05,
    max_depth=6,
    n_estimators=74,
    subsample=0.68,
)
```

```python
treemodel_rand.fit(X_train,y_train)
```

```python
y_pred = treemodel_rand.predict(X_test)
accuracy_score(y_test, y_pred)*100
```

```python
report = classification_report(y_test, y_pred)
print(report)
```
