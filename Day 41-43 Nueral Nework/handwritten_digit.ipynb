* Accuracy is way too much low so that we scale our features

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[0]
```

```python
X_train_flatten = X_train.reshape(len(X_train),28*28)
X_train_flatten.shape
```

```python
X_test_flatten = X_test.reshape(len(X_test),28*28)
X_test_flatten.shape
```

```python
X_train_flatten[0]
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, input_shape=(784,), activation='sigmoid'), # first one is outout layer, second one is input layer
])
model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
model.fit(X_train_flatten, y_train, epochs = 5)
```

```python
model.evaluate(X_test_flatten, y_test)
```

```python
plt.matshow(X_test[0])
```

```python
y_pred = model.predict(X_test_flatten)
y_pred[0]
```

```python
np.argmax(y_pred[0])
```

```python
y_pred_labels = [np.argmax(i) for i in y_pred]
y_pred_labels
```

# Both Values are same

```python
y_pred_labels[:5]
```

```python
y_test[:5]
```

```python
cm = tf.math.confusion_matrix(labels = y_test, predictions = y_pred_labels)
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot = True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

* Adding hidden layer

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100, input_shape=(784,), activation='relu'), # first one is outout layer, second one is input layer
    tf.keras.layers.Dense(10, activation='softmax'),
])
model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
model.fit(X_train_flatten, y_train, epochs = 5)
```

```python
model.evaluate(X_test_flatten, y_test)
```

```python
y_pred = model.predict(X_test_flatten)
y_pred[0]
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(100, activation='sigmoid'), # first one is outout layer, second one is input layer
    tf.keras.layers.Dense(10, activation='sigmoid') 
])
model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
model.fit(X_train, y_train, epochs = 10)
```
