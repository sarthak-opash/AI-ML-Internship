```python
import cv2 
from skimage.util import random_noise
from skimage import io, img_as_ubyte
from matplotlib import pyplot as plt
```

```python
img = io.imread('luffy.jfif')
```

```python
noisy = img_as_ubyte(random_noise(img, mode='s&p', amount=0.5))
denoised = cv2.medianBlur(noisy, 5)
```

```python
plt.imshow(noisy, cmap = 'gray')
```

```python
plt.imshow(denoised, cmap = 'gray')
```

```python
from skimage.color import rgb2gray
from skimage.color import rgb2hsv
from skimage.color import rgb2lab
```

```python
image = rgb2gray(img)
plt.imshow(image, cmap = 'gray')
```

```python
image = rgb2hsv(img)
plt.imshow(image)
```

```python
iamge = rgb2lab(img)
plt.imshow(iamge)
```

```python
from skimage import data
from skimage.filters import gaussian
from skimage.segmentation import active_contour, slic, mark_boundaries, felzenszwalb
```

```python
astronaut = data.astronaut()
gray_astronaut = rgb2gray(astronaut)
```

```python
plt.imshow(astronaut)
```

```python
plt.imshow(gray_astronaut)
```

```python
astronaut_segments = slic(astronaut, n_segments=200, compactness=5)

plt.subplot(1, 2, 1)
plt.imshow(astronaut)
plt.subplot(1, 2, 2)

plt.imshow(mark_boundaries(astronaut, astronaut_segments))
```

* label2rgb

```python
from skimage.color import label2rgb
plt.imshow(label2rgb(astronaut_segments,
                     astronaut,
                     kind = 'avg'))
```

```python
stronaut_segments = felzenszwalb(astronaut,
                                  scale = 2,
                                  sigma=5,
                                  min_size=100)

plt.subplot(1,2,1)
plt.imshow(astronaut)

plt.subplot(1,2,2)
plt.imshow(mark_boundaries(astronaut,
                           stronaut_segments))
```
