```python
from numpy.random import uniform
from numpy.random import randint
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
import pandas as pd 
import numpy as np 
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python

data1 = pd.read_csv('indian_engineering_student_placement.csv')
data2 = pd.read_csv('placement_targets.csv')
```

```python
data = pd.merge(data1, data2, on='Student_ID')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

cols = ['gender', 'branch', 'part_time_job', 'family_income_level', 'city_tier', 'internet_access', 'extracurricular_involvement', 'placement_status']

for col in cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
```

```python
data
```

```python
data['extracurricular_involvement'] = data['extracurricular_involvement'].fillna(data['extracurricular_involvement'].mean())
data
```

```python
data.isnull().sum()
```

```python
X = data.drop(columns=['placement_status', 'salary_lpa'])
y = data['placement_status']
```

```python
data.corr()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

# Parameter Selection For this data

```python
parameters = {

    'n_estimators': [20, 60, 100, 120],
    'max_features': [0.2, 0.6, 1.0],
    'max_depth': list(range(1, 11)),
    'max_samples': [0.5, 0.75, 1.0]
}
```

```python
from sklearn.ensemble import RandomForestClassifier

treemodel_grid = RandomForestClassifier()
cv=GridSearchCV(treemodel_grid,param_grid=parameters,cv=5,scoring='accuracy')
cv.fit(X_train,y_train)
```

```python
cv.best_params_
```

```python
cv.best_score_
```

```python
cv.best_index_
```

GridSearchCV Param

```python
treemodel_grid = RandomForestClassifier(
    max_features=1.0,
    max_depth=7,
    n_estimators=100,
    max_samples=0.5,
)
```

```python
treemodel_grid.fit(X_train,y_train)
```

```python
y_pred = treemodel_grid.predict(X_test)
accuracy_score(y_test, y_pred)*100
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

# Cross Validation

```python
from sklearn.model_selection import KFold, cross_val_score

kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
rft_score = cross_val_score(treemodel_grid, X, y, cv=kf)
rft_score


for i, result in enumerate(rft_score, 1):
    print(f"Fold {i}: {result * 100:.2f}%")
    
print(f'Mean Accuracy: {rft_score.mean()* 100:.2f}%')
```

```python
rft_score = cross_val_score(treemodel_grid, X_feature, y, cv=kf)
rft_score


for i, result in enumerate(rft_score, 1):
    print(f"Fold {i}: {result * 100:.2f}%")
    
print(f'Mean Accuracy: {rft_score.mean()* 100:.2f}%')
```
