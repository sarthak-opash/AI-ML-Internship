## Creating AgeGroup Table

 - This is created for to understand the which person is from which age standard
  - less than 12 = kid
  - 12 - 17 = young
  - 18 - 70 = adult
  - more than 70 = old

```python
df.loc[df["Age"] < 12, "AgeGroup"] = "Kid"
df.loc[(df["Age"] >= 12) & (df["Age"] < 18), "AgeGroup"] = "Young"
df.loc[(df["Age"] >= 18) & (df["Age"] < 70), "AgeGroup"] = "Adult"
df.loc[df["Age"] >= 70, "AgeGroup"] = "Old"
```

# Data Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("cleaned_dataset_for_pipeline.csv")
```

```python
X = df.drop(columns=["PassengerID", "Name", "FamilyName", "FullName", "Purchased"])
y = df["Purchased"]
```

```python
num_cols = ['Age','Salary','Children']
cat_cols = ['Gender','City','Education','AgeGroup']
```

```python
num_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
```

```python
cat_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])
```

```python
preprocessor = ColumnTransformer(transformers=[
    ("num", num_pipeline, num_cols),
    ("cat", cat_pipeline, cat_cols)
])
```

```python
model = Pipeline(steps=[
    ("preprocessing", preprocessor),
    ("classifier", LogisticRegression())
])
```

```python
model.fit(X, y)
```

```python
model.predict(X.head())
```
