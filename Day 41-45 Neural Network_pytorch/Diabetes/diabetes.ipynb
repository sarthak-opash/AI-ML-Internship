```python

import torch
from torch.utils.data import TensorDataset

# Convert pandas DataFrames/Series to numpy arrays before creating tensors
train_dataset = TensorDataset(
    torch.tensor(X_train.values, dtype=torch.float32), 
    torch.tensor(y_train.values, dtype=torch.float32)
)
len(train_dataset)
```

```python
test_dataset = TensorDataset(
    torch.tensor(X_test.values, dtype=torch.float32),
    torch.tensor(y_test.values, dtype=torch.float32)
)
len(test_dataset)
```

```python
X_test.shape
```

```python
from torch.utils.data import DataLoader
train_loader = DataLoader(train_dataset, batch_size = 32, shuffle = True)
test_loader = DataLoader(test_dataset, batch_size = 32 , shuffle = False)
```

```python

class MyNN(nn.Module):

    def __init__(self, num_feature):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(num_feature, 16),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(16, 8),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(8, 4),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(4, 1)
        )

    def forward(self, x):
        return self.model(x)
```

```python
epochs = 5000
learning_rate = 0.01
```

```python
model = MyNN(X_train.shape[1])
```

```python
criterion = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr = learning_rate)
```

```python
for epoch in range(epochs):
    model.train()
    epoch_loss = 0
    correct = 0
    total = 0
    
    for batch_feature, batch_label in train_loader:
        optimizer.zero_grad()
        
        y_pred = torch.sigmoid(model(batch_feature))
        loss = criterion(y_pred, batch_label.unsqueeze(1))
        loss.backward()
        optimizer.step()
        
        epoch_loss += loss.item()
        preds = (y_pred > 0.5).float()
        correct += (preds == batch_label.unsqueeze(1)).sum().item()
        total += batch_label.size(0)
        
    print(f"Epoch : {epoch+1} | Loss : {epoch_loss/len(train_loader):.4f} | Accuracy : {correct/total:.4f}")
```
