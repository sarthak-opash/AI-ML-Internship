```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

```

Creating DataFrame of the CSV File

```python
data = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")

data
```

```python
data.TotalCharges = pd.to_numeric(
    data.TotalCharges,
    errors = 'coerce'
)
data.isnull().sum()
```

```python
data.info()
```

```python
data = data.dropna()
df2 = data.iloc[:,1:].copy()

df2['Churn']=df2['Churn'].map({'Yes':1, 'No':0})

df_dummies = pd.get_dummies(df2, drop_first=True)
df_dummies.head()
```

```python
plt.figure(figsize=(15,8))
corr_churn = df_dummies.corr()[['Churn']].sort_values(by='Churn',ascending=False)
sns.heatmap(corr_churn, annot=True, cmap='coolwarm', center=0, cbar=True)
plt.title('Correlation of Features with Churn')
plt.show()
```
