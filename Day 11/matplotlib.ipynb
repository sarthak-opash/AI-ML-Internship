```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axes = np.array([2,5])
y_axes = np.array([4,8])
plt.plot(x_axes, y_axes, marker='o')
plt.title("Simple Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid()
plt.show()
```

```python
y_axis = ([1,8,3,5,77,88,45,212,2,3,0])
plt.plot(y_axis, marker='o')
plt.grid()
plt.show()
```

```python
x_axes1 = np.array([2,5])
y_axes1 = np.array([4,8])
y_axis1 = ([1,8,3,5,77,88,45,212,2,3,0])

plt.subplot(1,2,1)
plt.plot(x_axes1, y_axes1, marker='o')

plt.subplot(1,2,2)
plt.plot(y_axis1, marker='o')
```
