```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.set_style("whitegrid")
rs = np.random.RandomState(10)
data = rs.randn(100)
```

```python
sns.histplot(data)
```

```python
sns.histplot(data,kde=True,color='green')
```

```python
sns.lineplot(data)
```

```python
sns.displot(data)
```

```python
colors = (['r','g','b'])

sns.distplot(data, 
             kde=True, 
             hist=True, 
             bins=10,
             rug=True,
             hist_kws={"alpha": 0.3, "color": colors[0]},
             kde_kws={"color": colors[1], "lw": 2},
             rug_kws={"color": colors[2]}
             )
```

```python
import pandas as pd

sns.set_style("dark")
dataset = pd.read_csv("currency.csv")

sns.lineplot(
    x="Code",
      y="Symbol", 
      hue="Name",
      data=dataset)
      

plt.show()
```
