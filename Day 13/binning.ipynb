```python
import pandas as pd
```

```python
data = {
    'Age': [7, 45, 25, 60, 87, 40, 60, 48, 33, 52, 13],
}

df = pd.DataFrame(data)
print(df.head())
```

```python
bins = [0, 18, 35, 50, 65, 100]
labels = ['Teen', 'Young Adult', 'Adult', 'Middle Aged', 'Senior']

df_age_binned = pd.cut(df['Age'], bins=bins, labels=labels)
df['Age_Group'] = df_age_binned
print(df.head())
```
