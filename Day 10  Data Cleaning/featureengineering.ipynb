```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

Handling Numeric missing value via imputer

```python
data = {
    "Age": [20,21,25,23,np.nan,20],
    "Salary": [2000,3000,4000,5000,np.nan,np.nan]
}
df = pd.DataFrame(data)
df
```

```python
imputer = SimpleImputer(strategy='median') # strategy = most_recent , most_used, mean, median
df[['Age',"Salary"]] = imputer.fit_transform(df[['Age','Salary']])
print(df)
```
