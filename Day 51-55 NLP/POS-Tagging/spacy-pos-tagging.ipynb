```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp("Samsung is looking for new model launching in the market end of this year")
```

```python
doc.text
```

```python
doc[0]
```

```python
doc[0].tag_
```

```python
doc[0].pos_
```

```python
spacy.explain("NNP")
```

```python
for word in doc:
    print(word.text, "----->", word.pos_, word.tag_, spacy.explain(word.tag_))
```

```python
from spacy import displacy
```

```python
displacy.render(doc, style='dep', jupyter=True)
```
