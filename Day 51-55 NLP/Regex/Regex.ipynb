```python
import re
```

```python
chat1 = 'my number is 1234567890 and email is abc@xyz.com '
chat2 = 'here is is 123-456-7890, abc@xyz.com'
chat3 = "(123)-456-7890"
```

```python
pattern = '\d{10}|\(\d{3}\)-\d{3}-\d{4}'
matches = re.findall(pattern, chat3)
matches
```

```python
pattern_Email = '[a-z0-9A-Z]*@[a-z0-9A-Z]*\.[a-zA-z]'
matches = re.findall(pattern_Email, chat2)
matches
```
