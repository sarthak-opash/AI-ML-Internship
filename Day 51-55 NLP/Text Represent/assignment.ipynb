```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv("IMDB Dataset.csv")
```

```python
data.head()
```

# Preprocessing

```python
data['review'] = data['review'].str.lower()
```

```python
data
```

```python
import re
def remove_html_tag(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r"", text)
```

```python
text = "<html><body><h1>Title</h1><p>This is a paragraph.</p></body></html>"
remove_html_tag(text)
```

```python
data['review'] = data['review'].apply(remove_html_tag)
```

```python
data
```

# Finding total words in corpus

```python
total_words = data['review'].str.split().str.len().sum()
print(total_words)
```

# Bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
model = cv.fit_transform(data['review'])
```

```python
print(cv.vocabulary_)
```

```python
print(model)
```

```python
print(model[1].toarray())
```

# Bag of word N gram

```python
ngram = CountVectorizer(ngram_range=(1,4))
```

```python
model_n_gram = ngram.fit_transform(data['review'])
```

```python
model_n_gram
```

```python
print(ngram.get_feature_names_out())
```
