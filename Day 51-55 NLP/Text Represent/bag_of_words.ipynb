```python
import pandas as pd 
import numpy as np
```

```python
data = pd.DataFrame(
    {
        'text': ['people watch campusX', 'campusX watch campusX', 'people write comment', 'campusX write comment'], 'output': [1,1,0,0]
    }
)
```

```python
data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(data['text'])
```

```python
print(cv.vocabulary_) # it showing index of vocabulary (alpha)
```

```python
print(bow[1].toarray())
```

```python
print(bow[0].toarray())
```
